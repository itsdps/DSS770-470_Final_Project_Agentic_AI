```python
from dotenv import load_dotenv
from agents import Agent, Runner

load_dotenv()

# Define the agent with instructions
agent = Agent(
    name="History Tutor",
    instructions="You answer history questions clearly and concisely.",
    model="gpt-4o"
 )

# Run the agent asynchronously in a notebook cell
result = await Runner.run(agent, "When did the Roman Empire fall?")
print(result.final_output)
```

```python
from agents import Agent, Runner, function_tool

@function_tool
def get_weather(city: str):
    """Get the current weather for a city."""
    return f"The weather in {city} is sunny, 22°C."

weather_agent = Agent(
    name="Weather Assistant",
    instructions="Provide weather updates using the tools provided.",
    tools=[get_weather]
 )

result = await Runner.run(weather_agent, "What's the weather in London?")
print(result.final_output)
```

```python
# Create specialized agents
sales_agent = Agent(name="Sales Agent", instructions="Handle sales queries.")
support_agent = Agent(name="Support Agent", instructions="Handle tech support.")

# Define handoffs
sales_agent.handoffs = [support_agent]
support_agent.handoffs = [sales_agent]

# The main runner manages the transfers automatically
result = await Runner.run(sales_agent, "I need technical support with my order.")
```

```python
print(result.final_output)
```
